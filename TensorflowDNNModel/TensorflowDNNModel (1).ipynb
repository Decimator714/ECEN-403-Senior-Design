```
# Seperate true labels from training and testing datasets
training_features = xtraining[:,0:]
training_labels = ttraining[:,0:]
testing_features = xtesting[:,0:]
testing_labels = ttesting[:,0:]
```

```
# Import classes
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
```

```

#model = tf.keras.models.Sequential([
#    tf.keras.layers.Dense(64, activation='linear', input_shape=(8,)),
#    tf.keras.layers.Dense(32, activation='linear'),
#    tf.keras.layers.Dense(5, activation='linear')
#])

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_shape=[8]),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(1024, activation='relu'),
    tf.keras.layers.Dense(5)
])



# Print model summary
model.summary()
```

```
# Configure model training
# model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])

model.compile(loss='mse', optimizer='adam')

```

```
# Train model
#model.fit(training_features,training_labels,epochs=100,batch_size=32)

history = model.fit(training_features, training_labels, epochs=400, validation_data=(testing_features, testing_labels))
```

Test the model Accuracy

```
test_loss = model.evaluate(testing_features, testing_labels)
```

In machine learning, the goal of training a model is to minimize its loss function. The loss function measures the difference between the predicted output of the model and the true output for a given input. The loss function is typically defined as a mathematical function that takes as input the predicted output and the true output, and outputs a scalar value that represents the difference between the two.

The "loss" in the TensorFlow DNN model refers to the value of the loss function for the current batch of training data. This value is calculated using the predicted output of the model and the true output for the batch of training data.

The "val_loss" in the TensorFlow DNN model refers to the value of the loss function on the validation set. During training, a separate validation set is used to evaluate the performance of the model on data that it has not seen before. The "val_loss" is calculated using the predicted output of the model and the true output for the validation set.

Start the process of delpoying to c++ code

```
# Convert Keras model to a tflite model
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.OPTIMIZE_FOR_SIZE]
tflite_model = converter.convert()

open(tflite_model_name + '.tflite', 'wb').write(tflite_model)
```
