Settings

```
tflite_model_name = 'Fault_model'  # Will be given .tflite suffix
c_model_name = 'Fault_model'       # Will be given .h suffix
```

```
# Read CSV file as a dataframe with pandas
xdata = pd.read_csv('./Xdata_to_DNN.csv',header=None)
tdata = pd.read_csv('./Tdata_to_DNN.csv',header=None)
```

```
# Print top records from dataframe
xdata.head()
```

```
tdata.head()
```

```
# Convert datatype to float32
xdata = xdata.astype('float32')
tdata = tdata.astype('float32')
```

```
# Check modified data
xdata.head()
```

```
tdata.head()
```

```
# Convert dataframe to numpy arrary
xdata=xdata.to_numpy()
tdata=tdata.to_numpy()
```

```
# Print shape of numpy arrary
xdata.shape
```

```
tdata.shape
```

```
# Get training and testing dataset from numpy
xtraining = xdata[0:120]
xtesting = xdata[120:]
ttraining = tdata[0:120]
ttesting = tdata[120:]
```

```
# Print training and testing dataset shape
print(xtraining.shape)
print(xtesting.shape)
print(ttraining.shape)
print(ttesting.shape)
```

```
# Seperate true labels from training and testing datasets
training_features = xtraining[:,0:]
training_labels = ttraining[:,0:]
testing_features = xtesting[:,0:]
testing_labels = ttesting[:,0:]
```

```
# Print shape of input that training_feaures and true labels that is training_labels
print(training_features.shape)
print(training_labels.shape)
```

```
# Checking input shape to neural network
training_features[0].shape
```

```
training_labels[0].shape
```

```
# Import classes
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout
```

```

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(8,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(5)
])
   
```

```
# Configure model training
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
```

```
# Train model
model.fit(training_features,training_labels,epochs=200,batch_size=32)
```

Test the model Accuracy

```
loss, acc = model.evaluate(testing_features, testing_labels, verbose=0)
print('Accuracy: %.3f' % acc)
```

Start the process of delpoying to c++ code

```
# Convert Keras model to a tflite model
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.OPTIMIZE_FOR_SIZE]
tflite_model = converter.convert()

open(tflite_model_name + '.tflite', 'wb').write(tflite_model)
```

```
from tinymlgen import port
```

```
c_code = port(model, pretty_print=1)
print(c_code)
```
